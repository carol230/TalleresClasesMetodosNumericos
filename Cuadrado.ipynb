# CUADRADOS


```python
import matplotlib.pyplot as plt
import numpy as np

def funcion(x):
    return 0.25 * x + 5

x_values = np.linspace(-6, 6, 100)
y_values = funcion(x_values)

puntos_x = [-3, -1, 2]
puntos_y = [4.25, 4.75, 5.5]

plt.figure(figsize=(8, 6))

plt.plot(x_values, y_values, label='y = 0.25x + 5', color='blue')

plt.scatter(puntos_x, puntos_y, color='red', zorder=5, label='Puntos dados')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()
```
