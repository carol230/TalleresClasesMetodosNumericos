```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**3 - 3*x**2 + x - 1

# Create an array of x values
x = np.linspace(-2, 4, 400)
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^3 - 3x^2 + x - 1')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Plot of the function f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
from scipy.optimize import newton

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html
x0 = 100
points = []
x_r = newton(f, x0, args=(points,True),maxiter=100)  # Este no es método de Newton, es secante!
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="red", zorder=5)  # Add the root to the plot
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Extract x values from points
x_points = [point[0] for point in points]

# Plot the x points
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```

## x0=100

```python
from scipy.optimize import newton
import matplotlib.pyplot as plt
import numpy as np

x_points = []
y_points = []

def f(x: float, show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    x_points.append(x)       
    y_points.append(y)      
    if show:
        print(f"f({x}) = {y}")
    return y

def f_prime(x: float):
    return 3 * x**2 - 6 * x + 1

x0 = 100

x_r = newton(lambda x: f(x, True), x0, fprime=f_prime, maxiter=100)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(x_points, marker='o', color="orange", linestyle='-', label="x points")

plt.axhline(0, color="black", linewidth=0.5)  
plt.ylabel("f(x)")
plt.title("Plot of x points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
x = np.linspace(-10, 10, 400)
y = [f(val) for val in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="red", zorder=5, label="Raíz encontrada")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
